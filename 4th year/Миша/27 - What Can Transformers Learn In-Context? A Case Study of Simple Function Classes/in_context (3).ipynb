## exp1 extrapolation

```python
task = "linear_regression"
#task = "sparse_linear_regression"
#task = "decision_tree"
#task = "relu_2nn_regression"

run_id = "pretrained"

run_path = os.path.join(run_dir, task, run_id)
```

Будем проверять для модели предобученной на линейную регрессию

Batch size = 64

20 dimensional data

Нормальное распределение

Загружаем предобученную модель, создаем функции для сэмплинга точек и весов

```python
model, conf = get_model_from_run(run_path)

n_dims = conf.model.n_dims
batch_size = conf.training.batch_size

data_sampler = get_data_sampler(conf.training.data, n_dims)
task_sampler = get_task_sampler(
    conf.training.task,
    n_dims,
    batch_size,
    **conf.training.task_kwargs
)
```

Сэмплируем веса и точки и считаем значения функции

```python
task = task_sampler() 
xs = data_sampler.sample_xs(b_size=batch_size, n_points=conf.training.curriculum.points.end)
ys = task.evaluate(xs)
```

```python
xs.shape, ys.shape
```

Считаем предикты

```python
with torch.no_grad():
    pred = model(xs, ys)
```

Рисуем лосс (тут много не нужного кода)

```python
metric = task.get_metric()
loss = metric(pred, ys).numpy()

sparsity = conf.training.task_kwargs.sparsity if "sparsity" in conf.training.task_kwargs else None
baseline = {
    "linear_regression": n_dims,
    "sparse_linear_regression": sparsity,
    "relu_2nn_regression": n_dims,
    "decision_tree": 1,
}[conf.training.task]

plt.plot(loss.mean(axis=0), lw=2, label="Transformer")
plt.axhline(baseline, ls="--", color="gray", label="zero estimator")
plt.xlabel("# in-context examples")
plt.ylabel("squared error")
plt.legend()
plt.show()
```

Авторский пример с экстраполяцией, умножаем все инпуты на 2

---



$y = w_{1} x_{1} + w_{2} x_{2} + ... + w_{20} x_{20}$

```python
xs2 = 2 * xs
ys2 = task.evaluate(xs2)
with torch.no_grad():
    pred2 = model(xs2, ys2)
```

```python
loss2 = metric(pred2, ys2).numpy()

plt.plot(loss.mean(axis=0), lw=2, label="Transformer")
plt.plot(loss2.mean(axis=0) / 4, lw=2, label="Transformer on doubled inputs")
plt.axhline(baseline, ls="--", color="gray", label="zero estimator")
plt.xlabel("# in-context examples")
plt.ylabel("squared error")
plt.legend()
plt.show()
```

Видно что все неплохо работает, однако если умножать на 3, то становится значительно хуже

```python
xs3 = 3 * xs
ys3 = task.evaluate(xs3)
with torch.no_grad():
    pred3 = model(xs3, ys3)
```

```python
loss3 = metric(pred3, ys3).numpy()

plt.plot(loss.mean(axis=0), lw=2, label="Transformer")
plt.plot(loss3.mean(axis=0) / 9, lw=2, label="Transformer on tripled inputs")
plt.axhline(baseline, ls="--", color="gray", label="zero estimator")
plt.xlabel("# in-context examples")
plt.ylabel("squared error")
plt.legend()
plt.show()
```
