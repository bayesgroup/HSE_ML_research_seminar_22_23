```python
import os
import subprocess

datasets = ['Cora', 'Citeseer', 'Pubmed', 'ogbn-arxiv']
type_models = ['GAT'] #, 'GCN', 'GIN']
spasity = [0.05, 0.1, 0.3, 0.5, 0.8, 0.9, 0.95]
for tp in type_models:
    print('----------------------')
    print('type_model = ', tp)
    for ds in datasets:
        print('-_-_-_-_-_-_-_-_-')
        print('datasets = ', ds)
        for sp in spasity:
            print('-_-_-_-_-_-_-_-_-')
            print('spasity = ', sp)
            p = subprocess.run(
                ["python", "/content/UGTs-LoG/UGTs_GNN/main.py", 
                 "--command", "train", 
                 "--lr", "0.01", 
                 "--num_layers", "2", 
                 "--dim_hidden", "256", 
                 "--dataset", f"{ds}", 
                 "--heads", "1", 
                 "--train_mode", "score_only", 
                 "--linear_sparsity", f"{sp}", 
                 "--exp_name", "test10", 
                 "--epochs", "400", 
                 "--type_model", f"{tp}", 
                 "--weight_l1", "0", 
                 "--repeat_times", "1", 
                 "--sparse_decay", 
                 "--weight_decay", "0.0", 
                 "--random_seed", "100"],
                stdout=subprocess.PIPE,
            )
            print(p.stdout.decode())
        print('\n\n\n\n\n\n\n')

    print('\n\n\n\n\n\n\n\n')
```
